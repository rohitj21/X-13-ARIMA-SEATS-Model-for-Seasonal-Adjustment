```python
import matplotlib.pyplot as plt
import numpy as np

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 10))
time = np.arange(0, 200)

# Additive Outlier (AO)
data_ao = np.random.normal(0, 1, 200)
data_ao[100] += 10  # Introduce a sudden spike
axes[0].plot(time, data_ao, label="Additive Outlier")
axes[0].set_title("Additive Outlier (AO)")
axes[0].set_ylabel("Value")
axes[0].legend(loc="upper right")

# Level Shift (LS)
data_ls = np.random.normal(0, 1, 200)
data_ls[100:] += 5  # Introduce a level shift
axes[1].plot(time, data_ls, label="Level Shift", color="orange")
axes[1].set_title("Level Shift (LS)")
axes[1].set_ylabel("Value")
axes[1].legend(loc="upper right")

# Temporary Change (TC)
data_tc = np.random.normal(0, 1, 200)
decay = 0.9 ** np.arange(0, 100)  # Decay factor
data_tc[100:200] += 10 * decay  # Introduce a temporary change with decay
axes[2].plot(time, data_tc, label="Temporary Change", color="green")
axes[2].set_title("Temporary Change (TC)")
axes[2].set_xlabel("Time")
axes[2].set_ylabel("Value")
axes[2].legend(loc="upper right")

plt.tight_layout()
plt.savefig("Outliers_Types.png")  # Save the image to file
plt.show()
```
